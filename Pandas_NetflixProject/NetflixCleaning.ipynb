```python
# Selection using loc for specific rows and columns
print("Selection using loc:")
display(data.loc[0:5, ['title', 'country', 'type']])
```

```python
# Selection using iloc for specific rows and columns (e.g., first 5 rows, first 3 columns)
print("Selection using iloc:")
display(data.iloc[:5, :3])
```

```python
# Filter all 'Movie' type with release_year > 2020
filtered_data = data[(data['type'] == 'Movie') & (data['release_year'] > 2020)]
display(filtered_data.head())
```

```python
# Filter TV Shows from 'India' using query
india_tv_shows = data.query("country == 'India' and type == 'TV Show'")
display(india_tv_shows.head())
```

```python
# Group by 'country' and 'type', then count titles and find mean release year
agg_results = data.groupby(['country', 'type']).agg({
    'title': 'count',
    'release_year': 'mean'
}).rename(columns={'title': 'count', 'release_year': 'avg_release_year'})

display(agg_results.sort_values(by='count', ascending=False).head(10))
```

```python
# Count number of titles per rating
print("Top 5 countries by content count:")
display(data['country'].value_counts().head(5))
```

```python
# Pivot table showing count of Movies vs TV Shows per country (top 10 countries)
top_countries = data['country'].value_counts().head(10).index
pivot_df = data[data['country'].isin(top_countries)].pivot_table(
    index='country', 
    columns='type', 
    values='title', 
    aggfunc='count',
    fill_value=0
)
display(pivot_df)
```

```python
regions_data = pd.DataFrame({
    'country': ['United States', 'India', 'United Kingdom', 'Japan', 'South Korea'],
    'region': ['North America', 'South Asia', 'Europe', 'East Asia', 'East Asia']
})

# Perform a left merge
merged_data = data.merge(regions_data, on='country', how='left')
display(merged_data[['title', 'country', 'region']].head(10))
```

```python
# Concat: Split and recombine (stacking Movies and TV Shows vertically)
movies = data[data['type'] == 'Movie'].head(3)
tv_shows = data[data['type'] == 'TV Show'].head(3)
combined = pd.concat([movies, tv_shows], axis=0)
display(combined[['title', 'type']])
```

```python
# Use map for simple value replacement (e.g., shorter labels for types)
data['short_type'] = data['type'].map({'Movie': 'M', 'TV Show': 'TV'})

# Rename columns
data = data.rename(columns={'listed_in': 'genres'})

display(data[['title', 'short_type', 'genres']].head())
```

```python
# Melt: Reshape for analysis (taking title, type, and release_year and making them key-value pairs)
melted_df = data.head(5).melt(id_vars=['show_id'], value_vars=['title', 'type'], var_name='attribute', value_name='value')
display(melted_df)
```

```python
# str.contains: Find titles containing 'Action' or 'Drama' in genres (case insensitive)
drama_titles = data[data['genres'].str.contains('Drama', case=False, na=False)]
print("Sample Drama Titles:")
display(drama_titles[['title', 'genres']].head())

# str.split: Split genres by comma and expand (look at the first title's genres)
print("Split genres for first 5 titles:")
display(data['genres'].str.split(',').head(5))

# str.extract: Extract first word of title
data['first_word'] = data['title'].str.extract(r'^(\w+)')
display(data[['title', 'first_word']].head())
```

```python
# Convert date_added to actual datetime objects
data['date_added'] = pd.to_datetime(data['date_added'])

# Extract parts of the date
data['added_year'] = data['date_added'].dt.year
data['added_month'] = data['date_added'].dt.month_name()
data['added_day_name'] = data['date_added'].dt.day_name()

display(data[['title', 'date_added', 'added_year', 'added_month', 'added_day_name']].head())
```

```python
# describe: Statistical summary of numerical columns
print("Statistical Summary:")
display(data.describe())

# Mean and Median of release_year
print(f"Mean Release Year: {data['release_year'].mean():.2f}")
print(f"Median Release Year: {data['release_year'].median()}")

# Correlation (between numerical columns - here we only have release_year and added_year)
print("\nCorrelation Matrix:")
display(data[['release_year', 'added_year']].corr())
```

```python
# Count releases per year
yearly_counts = data.groupby('release_year').size().reset_index(name='count').sort_values('release_year')

# cumsum: Cumulative count of releases over time
yearly_counts['cumulative_total'] = yearly_counts['count'].cumsum()

# rolling: 5-year moving average of release counts
yearly_counts['moving_avg_5yr'] = yearly_counts['count'].rolling(window=5).mean()

# shift: Difference in count compared to previous year
yearly_counts['diff_previous_year'] = yearly_counts['count'] - yearly_counts['count'].shift(1)

display(yearly_counts.tail(10))
```

```python
# pd.cut: Bin release years into decades
bins = [1920, 1980, 1990, 2000, 2010, 2020, 2030]
labels = ['Pre-80s', '80s', '90s', '2000s', '2010s', '2020s']
data['decade'] = pd.cut(data['release_year'], bins=bins, labels=labels)
display(data[['title', 'release_year', 'decade']].head())

# pd.get_dummies: One-hot encoding for the 'short_type' column
dummies = pd.get_dummies(data['short_type'], prefix='type')
display(dummies.head())

# MultiIndex example: Grouping by Decade and Country
multi_indexed = data.groupby(['decade', 'country']).size().head(10)
print("MultiIndex Series (Decade, Country):")
print(multi_indexed)
```
