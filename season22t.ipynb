```python
import pandas as pd
import numpy as np
```

### What are Vectorized Operations

```python
a = np.array([1,2,3,4])
a * 4
```

#### Problem in vectorized operations in vanilla Python

```python
s = ['cat','mat',None,'rat']
try:
    [i.startswith('c') for i in s]
except AttributeError as e:
    print("AttributeError:", e)
```

#### How Pandas solves this issue? (String Accessor `.str`)

```python
s = pd.Series(['cat','mat',None,'rat'])
# string accessor
s.str.startswith('c')
```

### Importing Titanic Dataset

```python
df = pd.read_csv('titanic.csv')
df['Name']
```

### Common String Functions

#### 1. Case Modifications (upper, lower, capitalize, title)

```python
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
```

#### 2. Length of String (`.str.len()`)

```python
df['Name'][df['Name'].str.len() == 82].values[0]
```

#### 3. Stripping Whitespace (`.str.strip()`)

```python
"                   nitish                              ".strip()
```

```python
df['Name'].str.strip()
```

### Split and Get Elements

```python
df['lastname'] = df['Name'].str.split(',').str.get(0)
df.head()
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
df.head()
```

```python
df['title'].value_counts()
```

### Replace String Values

```python
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
df['title'].value_counts()
```

### Filtering Columns

#### 1. Starts with / Ends with

```python
df[df['firstname'].str.endswith('A')]
```

#### 2. Digit check / Alpha check

```python
df[df['firstname'].str.isdigit()]
```

### Sifting with Regular Expressions

#### 1. Substring Contains (Case Insensitive search)

```python
df[df['firstname'].str.contains('john',case=False)]
```

#### 2. Advanced Regex (Matching start and end vowels)

```python
df[df['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```

### Vectorized String Slicing

```python
df['Name'].str[::-1]
```
